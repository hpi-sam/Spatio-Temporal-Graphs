```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
causal_graph = nx.read_graphml('../causality.graphml')
```

```python
nx.draw(causal_graph)
```

```python
anomaly_nodes = [n for n in causal_graph.nodes(data=True) if n[1].get('anomaly')=='true']
anomaly_nodes
```

```python
remove_edges = [e for e in causal_graph.edges(data=True) if e[2].get('causal')!='true']
```

```python
causal_graph.remove_edges_from(remove_edges)
```

```python
nx.draw(causal_graph)
```

```python
anomaly_graph0 = nx.bfs_tree(causal_graph, anomaly_nodes[0][0])
nx.draw(anomaly_graph0)
```
